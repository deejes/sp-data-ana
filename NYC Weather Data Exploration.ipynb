```python
locations = pd.read_csv('gapi_locations.csv')
```

```python
locations = locations[['Criteria ID','Name']]
```

```python
locations.columns = ['ID','Name']
```

```python
locations = locations.set_index('ID')
```

```python
locations = locations.reset_index()
```

```python
locations
```

```python
def find_city_name(id):
    try:
        return locations.loc[locations['ID'] == id]["Name"].values[0]
    except:
        return "NA"
```
