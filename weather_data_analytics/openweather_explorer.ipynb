```python
city_dict = {4887398:'Chicago', 5128581:'NYC', 5809844:"Seattle"}
```

```python
df = pd.read_csv('open_weather_6cities.csv')
```

```python
df.dropna(axis=1, how='all', inplace=True) # drops all columns where all values are NaN
```

```python
df
```

```python
df = df[['dt','city_id','weather_id']]
```

```python
df = df.replace({'city_id':city_dict}) 
```

```python
df2 = df.pivot_table(index='city_id', columns='weather_id',values='dt', aggfunc='count')
```

```python
df2 = df2.fillna(0)
```

```python
df2.index.name = None
```

```python
df2.rename(columns=lambda x: (str(x))[0], inplace=True)
```

```python
df2 = df2.transpose()
```

```python
df2 = df2.groupby(df2.index).sum()
```
